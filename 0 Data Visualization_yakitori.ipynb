```python
### EDA ---------------------------------------
## EDA: Credit history
# Summary of status for each client-------------------------------
cre_group = cre_df.groupby('ID')
cre_pivot = cre_df.pivot(index='ID', columns = 'MONTHS_BALANCE',values='STATUS')
cre_pivot['open_month'] = cre_group['MONTHS_BALANCE'].min()
cre_pivot['end_month'] = cre_group['MONTHS_BALANCE'].max()
cre_pivot['window'] = cre_pivot['end_month'] - cre_pivot['open_month']+1
cre_columns = ['paid_off','pastdue_1-29','pastdue_30-59','pastdue_60-89','pastdue_90-119','pastdue_120-149','pastdue_over_150','no_loan']
cre_values = ['C','0','1','2','3','4','5','X']
for i in range(8):
    cre_pivot[cre_columns[i]] = cre_pivot[cre_pivot.iloc[:,0:61] == cre_values[i]].count(axis = 1)

cre_pivot
```

```python
# Status Ratio: how much do bad behaviors/good behavior count for each customer?
# Test assumptions: good customers and bad customers keep consistent pay-off behaviors. 
    # for bad behaviors, like past due over 2 months, the percentage is small and the pattern is repeated (percentage for bad behaviors for that customer is high)
    # for good customers, they keep paying off on time
cre_divisor = cre_pivot[['window','paid_off','pastdue_1-29','pastdue_30-59','pastdue_60-89','pastdue_90-119','pastdue_120-149','pastdue_over_150','no_loan']]
for i in range(8):
    cre_divisor[cre_columns[i]] = cre_divisor[cre_columns[i]]/cre_divisor['window']
display(cre_divisor)
display(cre_divisor.describe())
# we can see that most of the status are pastdue_1-29days, paid_off, and no_loan
# window variance is big. Thus, we need to consider the window length and corresponding credit score
    # how to define a fair credit score for each customer?
        # ratio
        # only look at a specific window for all customers
        # count of serious status
# Analysis:
    # paid_off: most people did not paid_off on time. Because median is 0, 75 percentile is 0.675676 and max is 1, customers who have paid_off record keep paid_off pattern, and seems to be "good" client
    # pastdue_1_29: this status is common
    # past_due_over_1_month: most people did not have past_due over 1 month.
    # past_over_150 only happens to some customers (these customers could be identified as "bad" client)
    # Test: if bad behavior occurs,is the client bad? 
    # no_loan: no_loan is not common as the median is 0.03. 

#%matplotlib inline
COL_NUM = 2
ROW_NUM = 4
fig, axes = plt.subplots(ROW_NUM, COL_NUM, figsize=(12,12))

for i in range(8): 
    ax = axes[int(i/COL_NUM), i%COL_NUM]
    sns.distplot(cre_divisor[cre_columns[i]], ax=ax)
    ax.set_title(cre_columns[i])

plt.tight_layout() 
```

```python
## Vintage analysis on credit score data (whole data to observe default pattern) ******************
# Vintage analysis is a method of evaluating the credit quality of a loan portfolio by analyzing net charge-offs in a given loan pool where the loans share the same origination period.

credit = pd.merge(cre_df, cre_pivot, on = 'ID', how = 'left')
credit = credit[credit['window'] > 20] # delete users whose windows less than 20
credit['month_on_book'] = credit['MONTHS_BALANCE'] - credit['open_month']
print(credit['ID'].nunique()) #21929 out of 36457
divisor = cre_pivot.groupby(['open_month']).agg({'window': ['count']}) 
divisor.reset_index(inplace = True)
divisor.columns = ['open_month','sta_sum']
cre_group = cre_df.groupby('ID')

def conduct_vintage(status, graph_name, credit = credit, divisor = divisor):
    credit['status'] = np.where((credit['STATUS'] == status), 1, 0) 
    credit['status'] = credit['status'].astype(np.int8)
    vintage = credit.groupby(['open_month','month_on_book']).agg({'ID': ['count']}) 
    vintage.reset_index(inplace = True)
    vintage.columns = ['open_month','month_on_book','sta_sum'] 
    vintage['due_count'] = np.nan
    vintage = vintage[['open_month','month_on_book','due_count']] 
    vintage = pd.merge(vintage,divisor, on = ['open_month'], how = 'left') 
    for open_m in range(-60,1): 
        ls = list()
        for mob in range(0,61): 
            due = list(credit[(credit['status'] == 1) & (credit['month_on_book'] == mob) & (credit['open_month'] == open_m)]['ID']) # 得到所有逾期用户的ID
            ls.extend(due) 
            vintage.loc[(vintage['month_on_book'] == mob) & (vintage['open_month'] == open_m), 'due_count'] = len(set(ls)) # 使用集合统计不重复ID的个数

    vintage['sta_rate']  = vintage['due_count'] / vintage['sta_sum'] 
    # draw graph
    vintage_wide = vintage.pivot(index = 'open_month',
                             columns = 'month_on_book',
                             values = 'sta_rate')
    lst = [i for i in range(0,61)]
    vintage_wide[lst].T.plot(legend = False, grid = True, 
                             title = 'past due over %s (for different open month with window over 20)'%(graph_name))
    plt.show()
    
    return vintage

    
v1 = conduct_vintage('1','30 days')

v2 = conduct_vintage('2','60 days')

v3 = conduct_vintage('3','90 days')

v4 = conduct_vintage('4','120 days')

v5 = conduct_vintage('5','150 days')


# Vintage Analysis:
# Most users' bad behavior will occur within 30 months. 
# To calculate a credit score for each client within same window, it is reasonable for us to use only the first 30 months data
```

```python
# First bad behavior happens at which month 
id_n = credit_sum['ID'].nunique()
def calculate_first_bad(command, c = credit, id_n = id_n): 
    credit = c.copy()
    credit['status'] = 0
    exec(command)
    #credit.loc[(credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1
    credit['month_on_book'] = credit['MONTHS_BALANCE'] - credit['open_month']
    minagg = pd.DataFrame(credit[credit['status'] == 1].groupby('ID')['month_on_book'].min())
    minagg['ID'] = minagg.index
    obslst = pd.DataFrame({'month_on_book':range(0,61), 'rate': None})
    lst = []
    for i in range(0,61):
        due = list(minagg[minagg['month_on_book']  == i]['ID'])
        lst.extend(due)
        obslst.loc[obslst['month_on_book'] == i, 'rate'] = len(set(lst)) / id_n 
    return obslst['rate']

command1 = "credit.loc[(credit['STATUS'] == '0') | (credit['STATUS'] == '1') | (credit['STATUS'] == '2') | (credit['STATUS'] == '3' )| (credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"   
command2 = "credit.loc[(credit['STATUS'] == '1') | (credit['STATUS'] == '2') | (credit['STATUS'] == '3' )| (credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"   
command3 = "credit.loc[(credit['STATUS'] == '2') | (credit['STATUS'] == '3' )| (credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"
command4 = "credit.loc[(credit['STATUS'] == '3' )| (credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"
command5 = "credit.loc[(credit['STATUS'] == '4' )| (credit['STATUS'] == '5'), 'status'] = 1"
command6 = "credit.loc[(credit['STATUS'] == '5'), 'status'] = 1"

obslst = pd.DataFrame({'past due more than 0 days': calculate_first_bad(command1),
                       'past due more than 30 days': calculate_first_bad(command2)
#                        'past due more than 60 days': calculate_first_bad(command3),
#                        'past due more than 90 days': calculate_first_bad(command4),
#                        'past due more than 120 days': calculate_first_bad(command5),
#                        'past due more than 150 days': calculate_first_bad(command6),
                        })

obslst.plot(grid = True, title = 'First Bad Behavior Observe Analysis')
plt.xlabel('Months on Books')
plt.ylabel('Cumulative %')
plt.show()

obslst = pd.DataFrame({#'past due more than 0 days': calculate_first_bad(command1),
                       #'past due more than 30 days': calculate_first_bad(command2),
                       'past due more than 60 days': calculate_first_bad(command3),
                       'past due more than 90 days': calculate_first_bad(command4),
                       'past due more than 120 days': calculate_first_bad(command5),
                       'past due more than 150 days': calculate_first_bad(command6),
                        })

obslst.plot(grid = True, title = 'First Bad Behavior Observe Analysis')
plt.xlabel('Months on Books')
plt.ylabel('Cumulative %')
plt.show()
```

```python
# Survivial Analysis
# pip install lifelines
from lifelines import KaplanMeierFitter
credit['truncate'] = credit.groupby('ID')['status'].transform('max')
truncate = credit.loc[credit['truncate']==1,]
not_trunc = credit.loc[credit['truncate']==0,]

dt1 = not_trunc.groupby('ID').agg({'month_on_book': ['max'],
                            'truncate':['max']
                           })
dt1.reset_index(inplace = True)
dt1.columns = ['ID','month_on_book','truncate']
dt2 = truncate.loc[truncate['status']==1,].groupby('ID').agg({'month_on_book': ['min'],
                            'truncate':['max']
                           })
dt2.reset_index(inplace = True)
dt2.columns = ['ID','month_on_book','truncate'] 
dt = dt1.append(dt2)

print(dt.head())

kmf = KaplanMeierFitter()
kmf.fit(dt['month_on_book'], event_observed=dt['truncate'], label='overdue')
kmf.plot()
```
